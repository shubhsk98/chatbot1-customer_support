#### Import Necessary Library

```python
# %% Imports
import os
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
```

#### Data Ingestion- Documentloaders

```python
## Reading a PDf File
loader=PyPDFLoader('Customer.pdf')
docs=loader.load()
docs
```

```python
type(docs[0])
```

##### Text Splitting from Documents- RecursiveCharacter Text Splitters

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)

final_documents = text_splitter.split_documents(docs)
final_documents
```

```python
print(final_documents[0])
#print(final_documents[1])
```

#### Embedding Techniques

Converting text into vectors

```python
from dotenv import load_dotenv

#load all the environment variables
load_dotenv()
```

```python
#os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY") - thisis used for openai
os.environ['HF_TOKEN']=os.getenv("HF_TOKEN")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
embeddings
```

```python
text="The app keeps crashing when I try to log in."
query_result=embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
doc_result = embeddings.embed_documents([text, "The app keeps crashing when I try to log in."])
doc_result[0]
```

```python
## Vector Embedding And Vector StoreDB
#from langchain_community.vectorstores import Chroma
#db=Chroma.from_documents(final_documents,embeddings)

from langchain_community.vectorstores import FAISS
db=FAISS.from_documents(final_documents,embeddings)
db
```

```python
### Retrieve the results from query vectorstore db
query="customer_query  intent  priority"
retrieved_results=db.similarity_search(query)
print(retrieved_results)
```

```python
retrieved_results[0].page_content
```

```python
## similarity Search With Score
query="customer_query  intent  priority"
docs = db.similarity_search_with_score(query)
docs
```

```python
## Saving to the disk = chrom db
#db=Chroma.from_documents(documents=final_documents,embedding=embeddings,persist_directory="./chroma_db")

### Saving And Loading
db.save_local("faiss_index")
```
